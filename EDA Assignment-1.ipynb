```python
#2.How does the death rate vary with age?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
age_bins = [0, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['0-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
death_rates = df.groupby('age_group')['DEATH_EVENT'].mean()
plt.figure(figsize=(10, 6))
death_rates.plot(kind='bar', color='red', alpha=0.7)
plt.title('Death Rate by Age Group Among Heart Failure Patients')
plt.xlabel('Age Group')
plt.ylabel('Death Rate')
plt.grid(True)
plt.show()
```

```python
#6.How does the prevalence of high blood pressure differ between male and female patients?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
gender_counts = df['sex'].value_counts()
high_bp_counts = df.groupby('sex')['high_blood_pressure'].sum()
percentage_male_high_bp = (high_bp_counts[1] / gender_counts[1]) * 100
percentage_female_high_bp = (high_bp_counts[0] / gender_counts[0]) * 100
print(f"Percentage of male patients with high blood pressure: {percentage_male_high_bp:.2f}%")
print(f"Percentage of female patients with high blood pressure: {percentage_female_high_bp:.2f}%")
plt.figure(figsize=(10, 6))
plt.bar(['Male', 'Female'], [percentage_male_high_bp, percentage_female_high_bp], color=['blue', 'pink'])
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage with High Blood Pressure')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
#7.what is the relationship between smoking habits and the occurrence of heart failure?
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
total_patients = len(df)
smokers_with_heart_failure = len(df[(df['smoking'] == 1) & (df['DEATH_EVENT'] == 1)])
non_smokers_with_heart_failure = len(df[(df['smoking'] == 0) & (df['DEATH_EVENT'] == 1)])
percentage_smokers_with_heart_failure = (smokers_with_heart_failure / total_patients) * 100
percentage_non_smokers_with_heart_failure = (non_smokers_with_heart_failure / total_patients) * 100
print(f"Percentage of smokers who had heart failure: {percentage_smokers_with_heart_failure:.2f}%")
print(f"Percentage of non-smokers who had heart failure: {percentage_non_smokers_with_heart_failure:.2f}%")
plt.figure(figsize=(8, 6))
plt.bar(['Smokers', 'Non-Smokers'], [percentage_smokers_with_heart_failure, percentage_non_smokers_with_heart_failure], color=['orange', 'green'])
plt.title('Occurrence of Heart Failure by Smoking History')
plt.xlabel('Smoking History')
plt.ylabel('Percentage with Heart Failure')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
#8.Are there any noticeable patterns in the distribution of death events across different age groups?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
age_bins = [0, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['0-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
death_event_percentages = df.groupby('age_group')['DEATH_EVENT'].mean() * 100
plt.figure(figsize=(10, 6))
plt.bar(age_labels, death_event_percentages, color='skyblue')
plt.title('Distribution of Death Events Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Death Events')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
#9.Is there any significant difference in election fraction between patients with and without diabetes
```

```python
import pandas as pd
import scipy.stats as stats
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
ef_diabetes = df[df['diabetes'] == 1]['ejection_fraction']
ef_no_diabetes = df[df['diabetes'] == 0]['ejection_fraction']
mean_ef_diabetes = ef_diabetes.mean()
mean_ef_no_diabetes = ef_no_diabetes.mean()
std_ef_diabetes = ef_diabetes.std()
std_ef_no_diabetes = ef_no_diabetes.std()
t_statistic,P_value = stats.ttest_ind(ef_diabetes,ef_no_diabetes,equal_var = False)
print(f"mean election fraction fro patients with diabetes:{mean_ef_diabetes:.2f}")
print(f"mean elcetion fraction for patients without diabetes:{mean_ef_no_diabetes:.2f}")
print(f"P-value from t-test :{P_value:.4f}")
alpha = 0.05
if P_value < alpha:
    print("there is a significant difference in election fraction between patients wiht and without diabetes.")
else:
    print("there is no significant difference in election fraction between patients with and without diabetes.")
```

```python
#10.How does the serum creative level vary between patients who survived and those who did not?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\heart_failure_clinical_records_dataset.csv')
survived = df[df['DEATH_EVENT'] == 0]['serum_creatinine']
not_survived = df[df['DEATH_EVENT'] == 1]['serum_creatinine']
plt.figure(figsize=(10, 6))
sns.histplot(survived, bins=20, kde=True, color='blue', label='Survived')
sns.histplot(not_survived, bins=20, kde=True, color='red', label='Not Survived')
plt.title('Distribution of Serum Creatinine Levels by Survival Status')
plt.xlabel('Serum Creatinine Level')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
